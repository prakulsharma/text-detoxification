```python
from datasets import load_dataset
import pandas as pd

# Loading English training dataset
dataset_e = load_dataset("s-nlp/paradetox")

# Loading Russian training dataset
dataset_r = load_dataset("s-nlp/ru_paradetox")

# Creating DataFrame for English training dataset
df_e = pd.DataFrame(dict(dataset_e)['train'])

# Creating DataFrame for Russian training dataset, combining train and validation splits
df_r = pd.concat([pd.DataFrame(dict(dataset_r)['train']), pd.DataFrame(dict(dataset_r)['validation'])])

# Displaying a few samples from each training DataFrame
print('English Training Dataset Samples:')
print(df_e.columns)
# print('\nRussian Training Dataset Samples:')
# print(df_r.head())
```
